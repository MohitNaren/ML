```python
import pandas as pd
satis_test = '/content/drive/MyDrive/ML_DATASETS/Copy of Santander Customer Satisfaction_test.csv'
df1 = pd.read_csv(satis_test)
df1.head()
```

```python
satis_train = '/content/drive/MyDrive/ML_DATASETS/Copy of Santander Customer Satisfaction_train.csv'
df2 = pd.read_csv(satis_train)
df2.head()
```

```python
df1.shape, df2.shape
```

```python
df1.isnull().sum()
df2.isnull().sum()
```

```python
df = pd.concat([df1,df2])
df.head()
```

```python
constant_f = [col for col in df.columns if df[col].nunique() == 1]
print(constant_f)
print(len(constant_f))
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

# Assuming df is your DataFrame and 'TARGET' is your target column
X = df.drop(columns=['TARGET'])
y = df['TARGET']

# Split into train and test sets
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
x_train.drop(labels = constant_f, axis=1, inplace=True)
x_test.drop(labels = constant_f, axis=1, inplace=True)
```

```python
import numpy as np
quasi_constant_f = []
for feature in x_train.columns:
    predominant = (x_train[feature].value_counts() / float(len(x_train))).sort_values(ascending=False).values[0]
    if predominant > 0.999:
        quasi_constant_f.append(feature)
print(f"Number of quasi-constant features in training set: {len(quasi_constant_f)}")
```

```python
x_train.drop(labels = quasi_constant_f, axis=1, inplace=True)
x_test.drop(labels = quasi_constant_f, axis=1, inplace=True)
```

```python
dups = []
for i in range(len(x_train.columns)):
    col1 = x_train.columns[i]
    for col2 in x_train.columns[i + 1:]:
        if x_train[col1].equals(x_train[col2]):
            dups.append(col2)
```

```python
x_train.drop(labels = dups, axis=1, inplace=True)
x_test.drop(labels = dups, axis=1, inplace=True)
```

```python
grp_f = []
cor_f = []
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
c_f = correlation(x_train,0.9)
c_f = set(c_f)
print(f"Number of correlated features in training set: {len(c_f)}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create correlation matrix
corr_matrix = x_train.corr()

# Plot heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, cmap='inferno', annot=False, linewidths=0.5)
plt.title("Correlation Matrix Heatmap (Before Dropping Features)")
plt.show()
```
